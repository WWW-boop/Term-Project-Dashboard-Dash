```python
from pycaret.regression import *
import pandas as pd

data = pd.read_csv('csv\pm25_new2.csv')
s = setup(data, target = 'PM25', session_id = 123)
```

```python
best = compare_models()
```

```python
preds = predict_model(best, data=data)
```

```python
preds.to_csv('csv\predicted_pm25.csv', index=False)
```

```python

evaluate_model(best)
```
